## Coverage & Indexing

---

### What are the different types of index coverage statuses in Google Search Console?

The Index Coverage report in GSC groups your site's URLs into four main categories, each with various sub-statuses:

1.  **Error:** Pages that couldn't be indexed. Google encountered a significant issue.
    *   **Examples:**
        *   `Server error (5xx)`: Your server returned an error.
        *   `Redirect error`: The redirect chain was too long, looped, or pointed to an invalid URL.
        *   `Submitted URL blocked by robots.txt`: You submitted a page for indexing (e.g., via sitemap) but your `robots.txt` file blocks Google from crawling it.
        *   `Submitted URL marked â€˜noindexâ€™`: You submitted a page for indexing, but it has a `noindex` meta tag or X-Robots-Tag HTTP header.
        *   `Soft 404`: The page returns a success code (200) to the browser but Google thinks it looks like an error page (e.g., "Not found" message but still a 200 status).
        *   `Not found (404)`: The URL returns a 404 error, indicating the page doesn't exist.

2.  **Valid with warnings:** Pages that are indexed but have some issues you might want to address.
    *   **Example:**
        *   `Indexed, though blocked by robots.txt`: The page was indexed *before* being blocked by `robots.txt`. Google won't re-crawl it, and it might eventually drop from the index.

3.  **Valid:** Pages that have been successfully indexed.
    *   **Examples:**
        *   `Submitted and indexed`: You submitted the URL (e.g., via sitemap) and Google indexed it.
        *   `Indexed, not submitted in sitemap`: Google found and indexed the page through other means (e.g., internal links, external links), but it's not in your sitemap.

4.  **Excluded:** Pages that were intentionally or unintentionally not indexed. This isn't necessarily bad; many are by design.
    *   **Examples:**
        *   `Excluded by â€˜noindexâ€™ tag`: You deliberately told Google not to index this page.
        *   `Blocked by robots.txt`: You're preventing Google from crawling these pages.
        *   `Page with redirect`: The URL redirects to another page, so this original URL isn't indexed.
        *   `Alternate page with proper canonical tag`: This page is a duplicate, and you've correctly specified the canonical version.
        *   `Crawled â€“ currently not indexed`: Google crawled the page but decided not to index it (more on this next).
        *   `Discovered â€“ currently not indexed`: Google knows the page exists but hasn't crawled it yet (more on this later).
        *   `Duplicate without user-selected canonical`: Google found duplicate pages and chose a canonical itself.
        *   `Duplicate, Google chose different canonical than user`: You specified a canonical, but Google chose a different one.

### How would you resolve â€œCrawled â€“ currently not indexedâ€ issues?

This status means Googlebot visited the page, understood its content, but decided it wasn't valuable enough or high enough quality to include in the index *at this time*.

**Reasons & Resolutions:**

1.  **Low-Quality Content:** The page might be thin, have duplicate content, or offer little unique value.
    *   **Resolution:**
        *   **Improve Content:** Add more substantial, unique, and helpful information. Make it significantly better than similar pages on the web.
        *   **Consolidate:** If you have multiple thin pages on similar topics, combine them into a comprehensive resource.
    *   **Example:** An FAQ page with very short, generic answers. *Resolution: Expand answers with detailed explanations, examples, and links to relevant resources.*

2.  **Duplicate Content (Internal or External):** Google might see it as too similar to other pages on your site or elsewhere.
    *   **Resolution:**
        *   **Canonicalize:** Ensure proper canonical tags point to the preferred version if it's an internal duplicate.
        *   **Rewrite/Differentiate:** If it's similar to external content, make yours substantially different and more valuable.
    *   **Example:** A product description copied verbatim from the manufacturer's site. *Resolution: Rewrite the description to be unique, focusing on user benefits and your brand's perspective.*

3.  **Poor User Experience:** High bounce rate, low time on page, or difficult navigation.
    *   **Resolution:** Improve page layout, readability, site speed, and mobile-friendliness.
    *   **Example:** A blog post with tiny font and intrusive pop-ups. *Resolution: Increase font size, remove aggressive pop-ups, and ensure fast loading.*

4.  **Not Enough Internal Links / Orphan Page:** The page might not have enough internal links pointing to it, signaling to Google that it's not important.
    *   **Resolution:** Add relevant internal links from other authoritative pages on your site.
    *   **Example:** A new service page that isn't linked from your main navigation or relevant blog posts. *Resolution: Add it to the main menu and link to it from existing, related content.*

5.  **Perceived as Not Valuable by Google's Algorithms:** This is a catch-all. Google might simply deem it not worth indexing given its current crawl budget and the overall quality signals of your site.
    *   **Resolution:**
        *   **Improve Overall Site Authority:** Work on E-E-A-T (Experience, Expertise, Authoritativeness, Trustworthiness) for your entire site.
        *   **Request Indexing (Sparingly):** After making improvements, use the URL Inspection tool in GSC to "Request Indexing."
        *   **Ensure it's in your Sitemap:** While not a guarantee, it helps.

**General Steps:**
1.  **Inspect URL:** Use the "URL Inspection" tool in GSC for the specific URL to get more details.
2.  **Review Content:** Honestly assess the page's quality and value.
3.  **Check Technicals:** Ensure no accidental `noindex` tags or `robots.txt` blocks (though this status usually means those aren't the primary issue).
4.  **Improve & Resubmit:** Make changes and then request indexing.

### What steps do you take when a page shows up as "Discovered - currently not indexed"?

This status means Google knows the URL exists (e.g., found it in a sitemap, or linked from another page) but hasn't crawled it yet. This can be due to several reasons:

1.  **Crawl Budget Issues:** Google allocates a certain "crawl budget" to each site. If your site is very large, has many low-quality pages, or server issues, Google might not get around to crawling new or less important pages quickly.
    *   **Resolution:**
        *   **Improve Site Speed:** Faster sites are easier to crawl.
        *   **Reduce Low-Value URLs:** `noindex` or `robots.txt` disallow unnecessary pages (e.g., filtered search results, old archives) to focus crawl budget on important content.
        *   **Fix Server Errors:** Ensure your server is responsive and not returning errors.
    *   **Example:** A large e-commerce site with thousands of product variations creating many URLs. *Resolution: Use canonicals for variations, improve server response time, and ensure the sitemap prioritizes key category and product pages.*

2.  **Perceived Low Importance/Authority:** Google might not prioritize crawling the page if it doesn't seem important.
    *   **Resolution:**
        *   **Internal Linking:** Add internal links from important, frequently crawled pages on your site to this newly discovered page. This signals its importance.
        *   **Sitemap Submission:** Ensure the URL is in your XML sitemap and the sitemap is submitted to GSC.
        *   **External Links (Long-term):** Gaining quality external links to the page or your site overall can improve its perceived authority.
    *   **Example:** A new blog post published on your site but not linked from the homepage or any other popular articles. *Resolution: Add a link to the new post from a relevant, high-traffic existing article or your homepage's "Recent Posts" section.*

3.  **Site Overload:** Google might temporarily postpone crawling if it thinks its crawler is overloading your server.
    *   **Resolution:** Check your server logs for high crawl rates around the time the page was discovered. If so, ensure your server can handle Googlebot's requests. You can (cautiously) adjust crawl rate in GSC settings if necessary, but improving server capacity is usually better.

4.  **Just a Matter of Time:** Sometimes, especially for new sites or pages, it just takes time for Google to get around to crawling.
    *   **Resolution:** Be patient, especially if you've already addressed the points above.

**General Steps:**
1.  **Inspect URL:** Use the "URL Inspection" tool to see if there's any more info. You can also click "Test Live URL" to see if Google *can* crawl it.
2.  **Improve Internal Linking:** This is often the most effective immediate step.
3.  **Check Sitemap:** Ensure the URL is included.
4.  **Request Indexing:** After ensuring the page is crawlable and valuable, use the "Request Indexing" feature in the URL Inspection tool. This can often nudge Google to crawl it sooner.

### How do you submit a new URL for indexing using Google Search Console?

There are two primary ways:

1.  **URL Inspection Tool (Recommended for individual or a few URLs):**
    *   Go to Google Search Console.
    *   In the top search bar, enter the full URL of the new page you want to index.
    *   GSC will retrieve data about the URL from the Google Index.
    *   If the page isn't indexed, you'll see a message like "URL is not on Google."
    *   Click the **"REQUEST INDEXING"** button.
    *   Google will run a quick live test to check for obvious indexing issues. If it passes, your request will be added to a crawl queue.
    *   **Example:** You've just published a new blog post `https://www.example.com/blog/new-awesome-post`. Paste this URL into the inspection tool and click "Request Indexing."

2.  **XML Sitemaps (Recommended for many URLs or ongoing site updates):**
    *   Create an XML sitemap (or ensure your CMS does this automatically). This is a file listing all the important URLs on your site.
    *   Include the new URL(s) in your sitemap.
    *   Submit your sitemap to Google Search Console:
        *   Go to "Sitemaps" under the "Index" section in the left-hand menu.
        *   Enter the URL of your sitemap file (e.g., `/sitemap.xml`) in the "Add a new sitemap" field.
        *   Click "SUBMIT."
    *   Google will periodically re-crawl your sitemap and discover new URLs listed there.
    *   **Example:** Your website `https://www.example.com` has a sitemap at `https://www.example.com/sitemap.xml`. When you publish 10 new product pages, you ensure they are all listed in this `sitemap.xml` file. Google will find them when it next processes the sitemap.

**Note:** Requesting indexing doesn't guarantee indexing or a specific ranking. The page still needs to meet Google's quality guidelines.

### What might cause a sudden drop in indexed pages, and how would you troubleshoot it?

A sudden drop in indexed pages is a serious issue. Here's how to approach it:

**Potential Causes:**

1.  **Accidental `robots.txt` Disallow:** A change to `robots.txt` might be blocking Googlebot from crawling large sections of your site.
    *   **Example:** Someone mistakenly adds `Disallow: /` to `robots.txt`.

2.  **Widespread `noindex` Tags:** `noindex` meta tags or X-Robots-Tag HTTP headers might have been accidentally applied to many pages (e.g., via a CMS plugin update or theme setting).
    *   **Example:** A new SEO plugin is installed, and its default setting adds `noindex` to all blog archive pages, which were previously indexed.

3.  **Server Issues:**
    *   **Site Down/Server Errors (5xx):** If Googlebot repeatedly encounters server errors when trying to crawl your site, it will start de-indexing pages.
    *   **Incorrect Redirects:** Faulty sitewide redirects (e.g., HTTP to HTTPS migration gone wrong) leading to loops or error pages.
    *   **Example:** Your hosting provider has a major outage, and your site is inaccessible for several hours/days.

4.  **Manual Action (Penalty):** Google might have applied a manual action for violating webmaster guidelines, which can lead to de-indexing.
    *   **Example:** A site engaged in aggressive link schemes gets a "Pure spam" manual action.

5.  **Security Issues (Hacking):** If your site is hacked and serving malware or spam, Google may de-index pages to protect users.
    *   **Example:** Your WordPress site gets compromised, and hackers inject spammy links and content.

6.  **Sitemap Issues:**
    *   **Sitemap Errors:** A corrupted or incorrectly formatted sitemap that GSC can no longer process.
    *   **Sitemap Removed:** If the sitemap was the primary way Google found many pages and it's removed or inaccessible.
    *   **Example:** An update to your sitemap generator plugin creates an empty or malformed `sitemap.xml`.

7.  **Domain/DNS Issues:** Problems with your domain registration or DNS settings could make your site inaccessible.
    *   **Example:** Your domain name expires and isn't renewed.

8.  **Google Algorithm Update (Less Common for *Sudden Drops* in *Indexed Count*):** While algorithm updates primarily affect rankings, a major quality algorithm might de-index very low-quality content, but usually, this is more gradual or tied to "Crawled - currently not indexed." A *sudden sharp drop* in the *total indexed count* usually points to technical or penalty issues.

**Troubleshooting Steps:**

1.  **Check GSC Coverage Report:**
    *   Go to GSC > Index > Coverage.
    *   Look at the "Error" and "Excluded" graphs. Is there a sudden spike in errors or a specific exclusion reason that coincides with the drop?
    *   Drill down into the specific error types (e.g., `Server error (5xx)`, `Blocked by robots.txt`, `Submitted URL marked â€˜noindexâ€™`).

2.  **Inspect `robots.txt`:**
    *   Go to your `yourdomain.com/robots.txt`. Does it look correct?
    *   Use GSC's `robots.txt` Tester (under "Crawl" in the old GSC, or check specific URLs with URL Inspection tool for blockage) to see if important sections are disallowed.

3.  **Check for `noindex` Tags:**
    *   Use the "URL Inspection" tool on a few affected (or formerly indexed) pages. Check the "Indexing" section for "Indexing allowed?" and "User-declared canonical."
    *   View the source code of a few pages: `Right-click > View Page Source`. Search for `<meta name="robots" content="noindex">`.
    *   Check HTTP Headers (using browser developer tools or online header checker) for `X-Robots-Tag: noindex`.

4.  **Review Server Logs & Site Availability:**
    *   Check if your site is accessible.
    *   Look at server logs for an increase in 5xx errors around the time of the drop.
    *   Use GSC > Settings > Crawl Stats to see if Googlebot encountered increased host availability issues.

5.  **Check GSC Manual Actions & Security Issues:**
    *   Go to GSC > Security & Manual Actions > Manual Actions.
    *   Go to GSC > Security & Manual Actions > Security Issues.
    *   Address any reported issues immediately.

6.  **Verify Sitemap Status:**
    *   Go to GSC > Index > Sitemaps.
    *   Check if your sitemap was processed successfully and when it was last read. Are there errors?

7.  **Check Domain & DNS:**
    *   Use a `whois` lookup to ensure your domain registration is active.
    *   Use a DNS checker tool (like `dnschecker.org`) to verify your DNS records are propagating correctly.

8.  **Google Search:**
    *   Perform a `site:yourdomain.com` search on Google. Does the number of results roughly match your expectations or previous indexed count? Are key pages missing?

By systematically going through these checks, you can usually pinpoint the cause of the sudden drop in indexed pages.

### What is the difference between Impressions and Clicks in GSC?

*   **Impressions:** An impression is counted each time one of your website's URLs appears in a search result page (SERP) that a user views. The user doesn't have to scroll to see your link for an impression to be counted, as long as it's on the page loaded.
    *   **Example:** If your page appears as the 5th result on a Google search page, and a user loads that page, it counts as one impression, even if the user doesn't scroll down to actually see your listing. If it appears again on page 2 and the user navigates to page 2, that's another impression.

*   **Clicks:** A click is counted when a user clicks on your link in the search results, taking them to your website.
    *   **Example:** If your page appears in search results (1 impression) and the user clicks on that link, it counts as one click.

**Key Difference:** Impressions measure visibility in search results; Clicks measure user engagement with those visible listings.

---

### How is Average Position calculated in Google Search Console?

The Average Position in GSC is the average ranking of your URL(s) for a specific query or set of queries. It's important to note:

1.  **It's an Average:** If your page ranks #2 for a query one day and #4 the next, the average position for that period would be #3.
2.  **Highest Position Counts:** If your URL appears multiple times on a single SERP (e.g., a regular organic link and an image pack link), GSC only counts the *highest* (numerically smallest) position for that impression.
3.  **Weighted by Impressions:** The average position for a query or page across multiple queries is weighted by the number of impressions each ranking position received. This means positions for queries with more impressions have a greater influence on the overall average.
4.  **Dynamic Nature:** Search results are highly dynamic and can vary based on user location, search history, device, and Google's personalization algorithms. So, the "average position" is truly an average across many different search contexts.
5.  **Not Always Intuitive:** If a page ranks for many long-tail keywords at lower positions but gets few impressions there, and ranks high for a few high-impression keywords, the average position will lean towards the high-ranking, high-impression keywords.

**Example:**
*   Query A: Your page ranks at position #1 and gets 1000 impressions.
*   Query B: Your page ranks at position #10 and gets 10 impressions.
*   The average position will be much closer to #1 than to #5.5 (the simple average of 1 and 10) because Query A has significantly more impressions.

GSC calculates it by taking the position of your site in search results for each impression. If your site appears at position 3 and position 7 for two different queries (or two different times for the same query), your average position would be (3+7)/2 = 5, assuming equal impressions. If impressions differ, it's a weighted average.

### How do you identify and improve low CTR (Click Through Rate) keywords?

CTR = (Clicks / Impressions) * 100. A low CTR means your page is appearing in search results, but users aren't clicking on it as much as they could be.

**Identifying Low CTR Keywords:**

1.  **Go to GSC Performance Report:** Navigate to Performance > Search results.
2.  **Enable Average CTR and Average Position:** Make sure these metrics are selected above the graph.
3.  **Filter by Position (Optional but Recommended):**
    *   Click "+ NEW" > "Position".
    *   Filter for positions where you expect a decent CTR (e.g., "Smaller than" 11 to see keywords on page 1, or even "Smaller than" 6 for top-of-page). Ranking on page 5 will naturally have a low CTR.
4.  **Filter by Impressions:**
    *   Click "+ NEW" > "Impressions".
    *   Filter for keywords with a significant number of impressions (e.g., "Greater than" 50 or 100). Improving CTR for keywords with very few impressions won't have a big impact.
5.  **Sort by CTR:** Click the "CTR" column header to sort from lowest to highest.

Now you have a list of keywords that:
*   Are ranking reasonably well.
*   Are getting a decent number of impressions.
*   Have a low click-through rate.

**Improving Low CTR:**

1.  **Analyze the SERP (Search Engine Results Page):**
    *   **Manually search for the keyword in an incognito window.** What do the results look like?
    *   **Competitor Titles & Descriptions:** What are your competitors doing? Are their titles more compelling? Do their descriptions better match user intent?
    *   **Rich Snippets:** Are competitors using schema markup to get rich snippets (stars, FAQs, product info) that make their listings stand out?
    *   **Featured Snippets/PAA:** Is there a featured snippet or "People Also Ask" box that might be satisfying user intent directly on the SERP?
    *   **Ads:** Are there many ads pushing organic results down?

2.  **Optimize Your Title Tag:**
    *   **Include the Keyword:** Naturally and prominently.
    *   **Compelling & Click-Worthy:** Use power words, numbers, questions, or address a pain point.
    *   **Match User Intent:** Does the title clearly promise what the user is looking for?
    *   **Optimal Length:** Keep it within Google's display limits (around 55-65 characters).
    *   **Example:**
        *   *Low CTR Title:* "Running Shoes Article"
        *   *Improved Title:* "10 Best Running Shoes for 2023 (Reviewed & Compared)"

3.  **Optimize Your Meta Description:**
    *   **Expand on the Title:** Provide more context and value.
    *   **Include Keywords (Naturally):** Google often bolds them in search results.
    *   **Strong Call to Action (CTA):** E.g., "Learn more," "Shop now," "Find out how."
    *   **Unique Selling Proposition (USP):** What makes your page/offering special?
    *   **Optimal Length:** Around 150-160 characters.
    *   **Example:**
        *   *Low CTR Description:* "This page talks about running shoes."
        *   *Improved Description:* "Discover the top 10 running shoes for road, trail, and stability. Read our expert reviews, compare features, and find your perfect pair today!"

4.  **Implement Schema Markup (Rich Snippets):**
    *   If applicable (reviews, products, FAQs, recipes, events, etc.), add structured data to your page. This can make your listing more visually appealing and informative.
    *   **Example:** Adding Review schema to a product page can display star ratings directly in the SERP.

5.  **Improve URL Structure (Minor Impact, but Good Practice):**
    *   A clean, descriptive URL can be a small positive signal.
    *   **Example:**
        *   *Less Ideal:* `example.com/page123.html`
        *   *Better:* `example.com/running-shoes/best-for-beginners`

6.  **Consider User Intent Mismatch:**
    *   Is your page *actually* what users are looking for when they type that query? If not, even the best title/description won't help much. You might need to create new content or significantly revise the existing page.
    *   **Example:** Your page is an informational guide about "types of mortgages," but users searching "apply for mortgage" have transactional intent. Your informational page will likely have low CTR for the transactional query.

7.  **Test and Iterate:** Make changes, wait a few weeks for GSC data to update, and then re-evaluate CTR. SEO is an ongoing process.

### How would you analyze a sudden drop in clicks despite stable impressions?

This scenario suggests your pages are still visible in search results (impressions are stable), but users are clicking on your listings less frequently.

**Analysis Steps:**

1.  **Verify Data Range & Filters:**
    *   In GSC Performance Report, ensure you're comparing the correct date ranges (e.g., "Compare" tab, last 28 days to previous period).
    *   Check if any filters (Query, Page, Country, Device) are accidentally applied that might skew the data.

2.  **Check CTR:** A drop in clicks with stable impressions *means* your CTR has dropped. Pinpoint which queries/pages are most affected.

3.  **Examine SERP Changes for Key Queries:**
    *   **New Competitors:** Has a new competitor with a very compelling title/description or rich snippet started ranking above or near you?
    *   **SERP Feature Changes:**
        *   **Lost Featured Snippet:** If you previously owned a featured snippet for a high-traffic query and lost it, clicks can plummet even if your organic blue link ranking is stable.
        *   **New SERP Features:** Has Google introduced new features like "People Also Ask" boxes, video carousels, or knowledge panels that are satisfying user intent directly on the SERP or pushing your result further down?
        *   **Ad Changes:** Has the number or prominence of ads increased for your key queries?
    *   **Example:** You rank #3 for "best coffee makers." Impressions are stable. Suddenly, Google adds a large "Shopping" ad block and a video carousel above your listing. Users might be clicking those instead.

4.  **Review Your Snippets (Title & Meta Description):**
    *   **Recent Changes:** Did you recently change your title tags or meta descriptions? The new versions might be less effective.
    *   **Google Rewriting Snippets:** Google sometimes rewrites titles and descriptions if it thinks its version is better or if yours is non-compliant/poor quality. Check what's actually appearing in SERPs for your key queries.
    *   **Example:** You A/B tested a new, "edgier" title tag for a key page, and it's actually deterring clicks.

5.  **Technical Issues Affecting Click-Through (Less Common, but Possible):**
    *   **Favicon Issues:** A missing or broken favicon can make your listing look less trustworthy or appealing. (Minor impact)
    *   **URL Changes:** If your URL structure changed and old URLs are redirecting, but the displayed URL in SERPs is momentarily confusing.

6.  **Seasonality or External Factors:**
    *   Is there a seasonal trend for the affected queries? (e.g., "Christmas gift ideas" will naturally see clicks drop after December).
    *   Major news events or trends shifting search behavior away from your topics.
    *   **Example:** A query like "summer vacation deals" will see clicks drop as summer ends, even if impressions for informational content around it remain somewhat stable initially.

7.  **Brand Perception Changes (Rare for Sudden Drops):**
    *   Negative PR or news about your brand could make users hesitant to click.

**Action Plan:**
1.  Identify the specific queries and pages with the largest CTR drop.
2.  Manually inspect the SERPs for these queries in incognito mode.
3.  If SERP changes are the cause, adapt: improve your snippets, try to win back featured snippets, or focus on other keywords.
4.  If your own snippet changes were detrimental, revert or try new variations.

### How can you use the Performance tab to identify keyword cannibalization issues?

Keyword cannibalization occurs when multiple pages on your own website compete for the same target keywords in Google search results. This can confuse search engines and dilute your ranking potential.

**Using GSC Performance Tab to Identify Cannibalization:**

1.  **Go to Performance > Search results.**
2.  **Click on the "QUERIES" tab.**
3.  **Filter for a Specific Query:**
    *   Click "+ NEW" > "Query..."
    *   Choose "Exact query" (or "Query containing" if you're looking at a broader topic).
    *   Enter the keyword you suspect might have cannibalization issues.
    *   Click "APPLY."
    *   **Example:** Filter for the query "best seo tools."

4.  **Switch to the "PAGES" Tab:** After applying the query filter, click on the "PAGES" tab below the graph.

5.  **Analyze the Results:**
    *   You will now see a list of all the pages on your site that are getting impressions (and clicks) *for that specific query*.
    *   **If you see multiple URLs from your site ranking prominently (e.g., both in the top 10 or 20) for the same exact query, this is a strong indicator of keyword cannibalization.**
    *   Pay attention to the impressions and clicks for each page. Sometimes one page gets the vast majority, which is less of an issue. But if several pages have significant impressions for the same term, it's problematic.

    *   **Example:** For the query "best seo tools," you see:
        *   `https://www.example.com/blog/best-seo-tools-2023` (500 impressions, 50 clicks)
        *   `https://www.example.com/resources/seo-software-guide` (300 impressions, 15 clicks)
        *   `https://www.example.com/seo-tools` (100 impressions, 2 clicks)
        This suggests these three pages are competing.

**Resolving Cannibalization (General Strategies):**

*   **Consolidate Content:** If pages are very similar, merge the content into the strongest page and 301 redirect the weaker ones to it.
*   **Differentiate Content:** If pages *should* be separate, refine their focus and target slightly different keyword variations or user intents. Optimize titles, headings, and content accordingly.
*   **Internal Linking:** Adjust internal links to point primarily to your preferred canonical page for that keyword.
*   **Canonical Tags:** Ensure the `rel="canonical"` tag on the "competing" pages points to your main target page for that keyword, if appropriate.
*   **Noindex (Carefully):** If a less important page is outranking your preferred page, you might consider `noindex` for the weaker page, but this is often a last resort.

### How would you identify pages with high impressions but no clicks?

These are pages that Google is showing in search results frequently, but users aren't clicking on them. This indicates a potential issue with your snippet, user intent mismatch, or very low ranking position (though "high impressions" usually means it's somewhat visible).

**Steps in GSC Performance Report:**

1.  **Go to Performance > Search results.**
2.  **Ensure Clicks, Impressions, and CTR are selected** above the graph.
3.  **Click on the "PAGES" tab.**
4.  **Filter for High Impressions:**
    *   Click "+ NEW" > "Impressions".
    *   Set a threshold, e.g., "Greater than" 100 (adjust based on your site's traffic).
    *   Click "APPLY."
5.  **Filter for Zero or Very Low Clicks:**
    *   Click "+ NEW" > "Clicks".
    *   Set "Equals" 0 (or "Smaller than" a very low number like 1 or 2 if you want to catch pages with negligible clicks).
    *   Click "APPLY."

    *Alternatively, after filtering for high impressions, you can simply sort the table by the "Clicks" column (ascending) or "CTR" column (ascending) to bring pages with 0 or very few clicks to the top.*

6.  **Analyze the List:** The resulting table shows pages that meet your criteria.
    *   **Check their average position.** If the position is very high (e.g., 50+), it's normal to have few clicks. The issue is more pressing for pages with high impressions and a somewhat reasonable average position (e.g., within the top 20-30).
    *   **Example:** A page `https://www.example.com/obscure-topic` has 500 impressions but 0 clicks, and its average position is 8.5. This is a prime candidate for investigation because it's on page 1 but isn't getting clicked.

**Next Steps After Identification:**
*   Investigate the SERP for the primary queries these pages are ranking for (use the method described above: click page, then queries tab).
*   Optimize title tags and meta descriptions.
*   Check for user intent mismatch.

### What actions can you take if a high-ranking page has a low CTR?

This means your page is doing well in terms of ranking (e.g., top 5-10), but users aren't enticed to click on your search result.

**Actions (similar to improving low CTR keywords, but page-focused):**

1.  **Analyze the SERP:**
    *   Search for the main keywords the page ranks for in an incognito window.
    *   **Competitor Snippets:** How do your competitors' titles and descriptions compare? Are they more compelling, emotional, or benefit-driven?
    *   **Rich Snippets:** Are competitors using schema markup (stars, FAQs, how-to, etc.) that makes their listing stand out?
    *   **Featured Snippet/PAA:** Is another site owning a prominent featured snippet or PAA box that's drawing clicks?
    *   **Ad Density:** Are ads pushing your organic result down, even if it's high-ranking?
    *   **Example:** Your page ranks #2 for "best budget laptop." The #1 result has review stars and a price range displayed. Your listing is plain text.

2.  **Rewrite Your Title Tag:**
    *   Make it more compelling and click-worthy. Use numbers, brackets, strong adjectives, address user intent directly.
    *   Ensure it's not truncated.
    *   **Example (before):** "Budget Laptops Page"
    *   **Example (after):** "Top 5 Best Budget Laptops Under $500 (2023 Reviews)"

3.  **Optimize Your Meta Description:**
    *   Make it a compelling summary that elaborates on the title and includes a call-to-action or unique selling proposition.
    *   Ensure it accurately reflects the page content.
    *   **Example (before):** "Find various budget laptops here. We list many affordable options."
    *   **Example (after):** "Looking for an affordable laptop? Our experts review the 5 best budget laptops under $500. Compare specs, pros & cons, and save big!"

4.  **Implement Structured Data (Schema Markup):**
    *   If applicable (reviews, products, FAQs, recipes, articles, etc.), add schema to enhance your SERP appearance.
    *   **Example:** For a recipe page, add Recipe schema to show cooking time, ratings, and an image.

5.  **Check for User Intent Mismatch:**
    *   Even if ranking high, if the page content doesn't *truly* satisfy what users are looking for with that query, they might scan the snippet and decide it's not for them.
    *   **Example:** Your page ranks for "how to fix leaky faucet" but it's a product page selling faucet parts. Users want a guide, not just parts initially. CTR will be low. *Solution: Create a guide, or ensure the snippet clearly indicates it's a parts page.*

6.  **Improve Page URL (if problematic):**
    *   While less impactful on CTR, a very messy or non-descriptive URL can sometimes look less trustworthy.

7.  **Consider the Date:**
    *   If your content is time-sensitive, ensure the title or description reflects its freshness (e.g., add the current year if appropriate). An old date can deter clicks.

8.  **A/B Test Snippets:**
    *   Change one element (title or description), wait for GSC data to update (a few weeks), and compare CTR. (Note: True A/B testing in SERPs is hard, this is more sequential testing).

### How can you track branded vs non-branded keyword traffic in GSC?

This helps you understand how much of your organic traffic comes from people specifically searching for your brand versus those discovering you through generic, non-branded terms.

**Steps in GSC Performance Report:**

1.  **Go to Performance > Search results.**
2.  **Tracking Branded Traffic:**
    *   Click "+ NEW" > "Query..."
    *   In the filter options:
        *   Choose "Queries containing" (or "Exact query" if your brand name is very unique and not part of other common phrases).
        *   Enter your brand name (and common variations or misspellings if necessary, using "OR" logic if your GSC version supports regex, or by adding multiple filters).
        *   `Filter: Query | Contains | YourBrandName`
        *   If you have multiple distinct brand terms, you might need to use the "Custom (regex)" option: `YourBrandName|YourOtherBrandTerm`
    *   Click "APPLY."
    *   The report will now show clicks, impressions, CTR, and position specifically for queries containing your brand name.
    *   **Example:** For a brand "Acme Widgets," filter for queries containing "acme widgets" or "acme."

3.  **Tracking Non-Branded Traffic:**
    *   Click "+ NEW" > "Query..."
    *   In the filter options:
        *   Choose "Queries not containing."
        *   Enter your brand name.
        *   `Filter: Query | Does not contain | YourBrandName`
    *   Click "APPLY."
    *   The report will now show metrics for all queries that *do not* include your brand name.

4.  **Compare:**
    *   Note the total clicks/impressions for branded queries.
    *   Remove the branded filter, apply the non-branded filter, and note those totals.
    *   You can use the "Compare" tab in GSC to look at these segments over different time periods.

**Why Track This?**
*   **Brand Strength:** High branded traffic indicates good brand awareness.
*   **SEO Effectiveness:** High non-branded traffic suggests your SEO efforts for generic keywords are successful.
*   **Campaign Impact:** See if marketing campaigns are increasing branded searches.
*   **Content Strategy:** Identify opportunities for more non-branded content to attract new users.

### What insights can you derive if mobile traffic CTR is lower than desktop CTR?

If mobile CTR is significantly lower than desktop CTR, especially for similar average positions, it points to potential issues with how your site appears or functions for mobile users in SERPs, or the user intent on mobile.

**Potential Insights & Causes:**

1.  **Poor Mobile Snippet Appearance:**
    *   **Title/Description Truncation:** Titles and descriptions have different display lengths on mobile. If yours are too long, they might be awkwardly truncated on mobile, making them less compelling.
    *   **Different SERP Layout:** Mobile SERPs can be more crowded with features (PAA, video carousels, etc.) that push organic results down or make them less prominent.
    *   **Action:** Check your snippets on mobile SERPs. Optimize for mobile length and appeal.

2.  **User Intent Differences on Mobile:**
    *   Mobile users are often looking for quick answers, local information ("near me" searches), or tasks they can complete on the go. If your page content or snippet doesn't align with this mobile-first intent for certain queries, CTR might be lower.
    *   **Example:** A query like "pizza delivery" is overwhelmingly mobile. If your desktop-focused page about "gourmet pizza recipes" ranks, mobile CTR will be low because users want to order, not cook.
    *   **Action:** Analyze queries with low mobile CTR. Does your page satisfy the likely mobile intent?

3.  **Site's Mobile User Experience (Post-Click Fear, Indirectly Affecting CTR):**
    *   While technically a post-click factor, if users have had bad mobile experiences with your site before, they might be less likely to click your result again.
    *   Google might also subtly down-rank or show less prominence to sites it knows offer a poor mobile experience, indirectly affecting CTR.
    *   **Action:** Ensure your site is mobile-friendly (use Google's Mobile-Friendly Test), loads fast on mobile (PageSpeed Insights), and has easy navigation.

4.  **Slower Mobile Page Load Speed:**
    *   If users know your site is slow on mobile, they might avoid clicking. Google also prioritizes faster sites.
    *   **Action:** Optimize mobile page speed.

5.  **Competition More Mobile-Optimized:**
    *   Your competitors might have snippets or landing pages that are better optimized for mobile users, drawing clicks away from you.
    *   **Action:** Analyze competitor snippets and mobile landing pages for relevant queries.

6.  **Different Ranking Positions:**
    *   Crucially, **check if the average position is significantly different between mobile and desktop for the same queries/pages.** If your mobile ranking is much lower, that's the primary reason for lower mobile CTR.
    *   **Action:** If mobile rankings are worse, focus on mobile SEO: mobile-friendliness, page speed, mobile-first indexing considerations.

**How to Investigate in GSC:**
1.  Performance Report > Filter by "Device: Mobile" vs. "Device: Desktop".
2.  Compare CTR for top queries and pages.
3.  Pay close attention to the "Average Position" for mobile vs. desktop. If mobile position is much worse, that's often the main culprit.

### What would you do if impressions from a key target country suddenly drop?

A sudden drop in impressions from a specific target country is alarming and needs immediate investigation.

**Troubleshooting Steps:**

1.  **Verify in GSC:**
    *   Go to Performance > Search results.
    *   Click the "COUNTRIES" tab.
    *   Select the "Compare" date range option to compare the problematic period with a previous, normal period.
    *   Confirm the drop is specific to that country and not a sitewide issue.
    *   **Example:** Comparing "Last 7 days" to "Previous 7 days" and seeing a 90% drop in impressions from "Germany" while "USA" remains stable.

2.  **Check `hreflang` Implementation (if applicable):**
    *   If you use `hreflang` tags to target different countries/languages, errors here are a common cause.
    *   **Issues:** Incorrect country/language codes, broken links in `hreflang` tags, missing return tags (e.g., DE page links to EN, but EN doesn't link back to DE), `hreflang` pointing to non-canonical URLs.
    *   **Tools:** Use GSC's old "International Targeting" report (if still accessible for your property type) or third-party `hreflang` validators. Manually inspect `hreflang` tags on key pages.
    *   **Example:** Your `hreflang` tag for German users `de-DE` mistakenly points to a 404 page, or your English page `en-US` doesn't have a return `hreflang` tag pointing back to the German version.

3.  **Server/CDN Issues Specific to that Region:**
    *   Is your server or Content Delivery Network (CDN) having issues serving content to that specific geographical region?
    *   **Action:** Check CDN status pages. Use VPNs or online tools (e.g., GeoPeeker, Uptrends' "Website Uptime Test" from different locations) to test site accessibility and speed *from that country*.
    *   **Example:** Your CDN has an outage affecting its European PoPs (Points of Presence), making your site slow or inaccessible in Germany.

4.  **Local Penalties or Algorithm Updates:**
    *   While less common for *just impressions* to drop without clicks/rankings also tanking, Google sometimes rolls out algorithm updates or applies penalties that might affect specific regions differently, especially if there are local search quality guidelines.
    *   **Action:** Search for news about Google algorithm updates that might have a regional impact. Check GSC for Manual Actions (though these are usually site-wide).

5.  **Competitor Changes in that Region:**
    *   Have strong local competitors significantly improved their SEO or launched major campaigns in that country, pushing your results down for many queries?
    *   **Action:** Use a VPN to simulate searching from that country and analyze the SERPs for your main keywords.

6.  **Legal or Technical Restrictions:**
    *   Are there any new legal requirements or technical restrictions (e.g., GDPR compliance issues if not handled correctly, government-level blocking) in that country affecting your site's visibility?
    *   **Example:** Your site is found non-compliant with a new data privacy law in that country, leading to reduced visibility or blocking.

7.  **Accidental `robots.txt` or `noindex` for Regional Content:**
    *   If you have country-specific subdomains or subdirectories, ensure they haven't been accidentally blocked or `noindexed`.
    *   **Example:** A `robots.txt` rule `Disallow: /de/` accidentally gets added, blocking your German subdirectory.

8.  **Check Google Trends / Local Search Demand:**
    *   Is there a genuine drop in search interest for your topics/products in that country? (Less likely to be "sudden" unless tied to a major event).
    *   **Action:** Use Google Trends, filtered by that country.

**Immediate Focus:** `hreflang` errors and regional accessibility (CDN/server issues) are often the most common culprits for sudden, country-specific impression drops.

### How do you check which websites link to your site using Google Search Console?

GSC provides a report on links to your site:

1.  **Go to Google Search Console.**
2.  In the left-hand navigation pane, scroll down to the "Links" section and click on it.
3.  The Links report has several sections:
    *   **External links > Top linking sites:** This is the primary section you're looking for. It lists the external domains that link to your website most frequently, ordered by the number of linking pages from that site.
    *   You can click on any domain in this list to see the specific pages on your site that it links to.
    *   Click "MORE" under "Top linking sites" to see a more extensive list.

**What you'll see:**
*   **Linking site:** The domain of the website linking to you.
*   **Linking pages:** The number of individual pages on that external site that link to your site.
*   **Target pages:** The number of unique pages on your site that are being linked to by that external site.

**Example:**
You navigate to Links > Top linking sites. You might see:
*   `someblog.com` | Linking pages: 15 | Target pages: 3
*   `industrynews.net` | Linking pages: 8 | Target pages: 8
*   `partnerwebsite.org` | Linking pages: 5 | Target pages: 1

Clicking on `someblog.com` would then show you which of your 3 target pages are being linked to from those 15 pages on `someblog.com`.

### What is the difference between Top Linked Pages and Top Linking Sites in GSC?

These are two distinct reports within the "Links" section of GSC:

1.  **Top Linked Pages (under "External links"):**
    *   **What it shows:** This report lists *your own website's pages* that have received the most backlinks from external websites. It tells you which of *your* content is most popular or authoritative in terms of attracting links.
    *   **Order:** Usually sorted by the number of incoming links or linking sites.
    *   **Usefulness:** Helps identify your most valuable content assets ("link magnets"), understand what kind of content attracts links, and find pages that could benefit from internal linking to pass link equity.
    *   **Example:**
        *   `https://www.yoursite.com/ultimate-guide-topic-x` (150 Incoming links, 25 Linking sites)
        *   `https://www.yoursite.com/popular-tool` (120 Incoming links, 18 Linking sites)
        *   `https://www.yoursite.com/blog/viral-post` (90 Incoming links, 40 Linking sites)

2.  **Top Linking Sites (under "External links"):**
    *   **What it shows:** This report lists the *external domains* (other websites) that link to your website most frequently or have the most pages linking to your site. It tells you *who* is linking to you.
    *   **Order:** Usually sorted by the number of linking pages from that external site.
    *   **Usefulness:** Helps identify your biggest "fans" or referring domains, assess the quality and relevance of sites linking to you, and spot potential link networks or spam if unexpected sites appear.
    *   **Example:**
        *   `influentialblog.com` (Links from 50 of their pages to yours)
        *   `industryforum.net` (Links from 30 of their pages to yours)
        *   `partnercompany.com` (Links from 10 of their pages to yours)

**In a Nutshell:**
*   **Top Linked Pages:** *Your* most linked-to content.
*   **Top Linking Sites:** *External websites* that link to you the most.

### How do you evaluate whether backlinks shown in GSC are helping your rankings?

GSC shows you *that* links exist, but evaluating their quality and impact requires more analysis, often supplemented by third-party tools. Not all links are good.

**Evaluation Steps:**

1.  **Relevance of the Linking Site:**
    *   **Check in GSC:** Go to Links > Top Linking Sites. Visit the linking domains.
    *   **Question:** Is the linking website topically relevant to your website or the specific page it's linking to? A link from a site in a completely unrelated niche is generally less valuable (and could even be spammy) than one from a relevant industry site.
    *   **Example:** If you sell pet supplies, a link from a popular dog training blog is highly relevant and likely helpful. A link from a Russian casino site is irrelevant and potentially harmful.

2.  **Authority/Trustworthiness of the Linking Site:**
    *   **Manual Check:** Does the site look professional and credible? Is its content well-written and original? Does it have a clear "About Us" and contact information?
    *   **Third-Party Metrics (Use with caution, as proxies):** Tools like Ahrefs (Domain Rating - DR), Moz (Domain Authority - DA), SEMrush (Authority Score - AS) provide metrics to estimate a site's authority. Higher scores generally indicate more authoritative sites.
    *   **Example:** A link from a major news outlet (e.g., nytimes.com) or a .gov/.edu site typically carries more weight than a link from a brand new, unknown blog.

3.  **Context of the Link & Anchor Text:**
    *   **In GSC:** Click on a "Top Linking Site," then click on one of your "Target pages" to see (some of) the linking pages from that external site (GSC doesn't always show all linking pages or anchor text directly in this view easily, third-party tools are better for this).
    *   **Visit the Linking Page:** Where on the page is your link? Is it editorially placed within relevant content? Or is it in a footer, sidebar, or a long list of unrelated links (which is less valuable)?
    *   **Anchor Text:** What text is used for the link? Is it relevant to your linked page's topic? Natural, varied anchor text is good. Over-optimized, exact-match anchor text from many low-quality sites can be a red flag.
    *   **Example (Good):** A food blog writes an article "10 Must-Try Pasta Sauces" and links to your "Authentic Marinara Recipe" page with the anchor text "this delicious marinara recipe."
    *   **Example (Potentially Bad):** A list of 100 random links in a blogroll, where your site "example.com" is linked with the anchor "buy cheap widgets."

4.  **Link Type (Follow vs. Nofollow):**
    *   GSC doesn't explicitly differentiate follow/nofollow in its main link reports. You'd need to check the linking page's source code or use browser extensions/third-party tools.
    *   `Follow` links (the default) pass PageRank ("link equity") and are generally more impactful for SEO.
    *   `Nofollow`, `sponsored`, `ugc` links typically don't pass PageRank directly, but can still drive referral traffic and contribute to brand visibility. Google may choose to treat some `nofollow` links as hints for ranking.
    *   **A healthy backlink profile has a mix, but you want a good number of relevant `follow` links.**

5.  **Referral Traffic:**
    *   Check Google Analytics (Acquisition > All Traffic > Referrals). Is the linking site sending actual visitors to your website? Links that drive traffic are often a sign of a good, editorially placed link.
    *   **Example:** If `toplinkingblog.com` shows up in GSC and also sends 50 visitors a month via Google Analytics, that's a positive signal.

6.  **Potential for Harm (Negative SEO):**
    *   Are there many links from very low-quality, spammy, or irrelevant sites? These could potentially harm your rankings.
    *   **Signs of bad links:** Foreign language sites (if you don't target that language), adult/gambling sites (if irrelevant), sites with "SEO" or "links" in the domain, sites with very thin or duplicate content.
    *   If you find a pattern of harmful links, you might consider using the Disavow Tool in GSC (use with extreme caution).

**Overall:** Look for a natural, diverse backlink profile from relevant, authoritative websites with contextually placed links. A few high-quality links are far more valuable than thousands of low-quality ones.

### How does Google notify you of a manual penalty, and how would you fix it?

**Notification:**

1.  **Google Search Console Message:**
    *   The primary notification method is through Google Search Console. You'll see a message in the "Security & Manual Actions" section, specifically under "Manual Actions."
    *   The message will describe the type of manual action (e.g., "Unnatural links to your site," "Thin content with little or no added value," "Pure spam").
    *   It will often specify whether the action affects the whole site or just specific parts/pages.

2.  **Email Notification:**
    *   GSC will also typically send an email to the verified site owners and users associated with the GSC property, alerting them to the new manual action message.

**Fixing a Manual Penalty (General Process):**

The exact steps depend on the *type* of manual action, but the general workflow is:

1.  **Understand the Reason:**
    *   Carefully read the manual action message in GSC. It will provide clues (and sometimes examples) of what Google found to be in violation of its Webmaster Guidelines (now Search Essentials).
    *   Consult Google's documentation on manual actions for detailed explanations of each type.

2.  **Identify and Rectify the Violating Issues:**
    *   **Example: "Unnatural links to your site"**
        *   **Identify:** Use GSC's Links report and third-party backlink tools (Ahrefs, Moz, SEMrush) to audit your backlink profile. Look for paid links, links from spammy directories, link exchange schemes, comment spam links, etc.
        *   **Rectify:**
            *   Contact webmasters of the linking sites and request removal of the offending links. Keep records of your outreach.
            *   For links you cannot get removed, use Google's Disavow Tool to tell Google to ignore them. Create a disavow file listing the domains or URLs to disavow. (Use this tool with extreme caution and only after trying to get links removed).
    *   **Example: "Thin content with little or no added value"**
        *   **Identify:** Review your site's content, especially pages with low word count, duplicate content, auto-generated content, or pages that offer little unique information (e.g., thin affiliate pages). GSC might provide some example URLs.
        *   **Rectify:**
            *   Improve the content significantly: add unique value, in-depth information, original research, or insights.
            *   Consolidate multiple thin pages into one comprehensive resource.
            *   Remove pages that cannot be improved and offer no value.
    *   **Example: "Hacked content"** (This is also a security issue)
        *   **Identify & Rectify:** Follow steps to clean hacked content (see next question).

3.  **Document Your Efforts:**
    *   Keep a detailed record of all the actions you took to fix the issues (e.g., spreadsheets of links removed/disavowed, lists of pages improved/removed, dates of changes).

4.  **Submit a Reconsideration Request:**
    *   Once you're confident you've addressed *all* the issues cited in the manual action (and any other potential violations), go to the "Manual Actions" report in GSC.
    *   There will be an option to "Request Review."
    *   Write a clear, concise, and honest reconsideration request.
        *   Acknowledge the issue.
        *   Explain what caused it (if you know).
        *   Detail the specific steps you took to fix it. (Refer to your documentation).
        *   Assure Google that you understand the guidelines and have taken steps to prevent recurrence.
        *   Be polite and professional.
    *   **Example for Unnatural Links:** "Dear Google Search Quality Team, We received a manual action for unnatural links. We have audited our backlink profile and identified [number] links violating guidelines, primarily from [types of sites]. We successfully removed [number] links by contacting webmasters (see attached spreadsheet of outreach) and have disavowed the remaining [number] links (disavow file submitted on [date]). We are committed to building links naturally moving forward..."

5.  **Wait for Google's Response:**
    *   Google will review your request. This can take several days to a few weeks (or longer).
    *   You'll receive a response in GSC (and via email) informing you if the request was approved (penalty revoked) or rejected (more work needed).
    *   If rejected, they may provide additional feedback. Go back to step 1 and try again.

### If GSC shows a security issue like malware or hacked content, what steps should you take?

Security issues are critical and need immediate attention to protect your users and your site's reputation.

**Notification:**
*   GSC will display a warning in the "Security & Manual Actions" section, under "Security Issues."
*   Common types: Malware, Hacked Content (e.g., spam injection), Deceptive Pages (Phishing).
*   You'll also likely receive an email notification.
*   Google may show warnings to users in search results or browsers when they try to visit your site.

**Steps to Take:**

1.  **Isolate/Contain (If Possible/Necessary):**
    *   **Take the site offline (maintenance mode):** This can prevent further harm to users or stop the spread of malware if it's actively compromising visitors. This isn't always the first step, but consider it for severe cases.
    *   **Contact your hosting provider:** Inform them of the issue. They might have tools or procedures to help, or they might take your site offline themselves if it's affecting other sites on shared hosting.

2.  **Identify the Hack/Malware:**
    *   **GSC Details:** The Security Issues report in GSC will often provide sample URLs that are affected and sometimes details about the type of hack.
    *   **Use Security Scanners:**
        *   Run tools like Sucuri SiteCheck, Wordfence (if WordPress), Google's Safe Browsing site status checker.
        *   Your hosting provider might offer a malware scanner.
    *   **Check Server Logs:** Look for suspicious activity, unfamiliar IP addresses accessing files, or unusual requests.
    *   **Examine Files:** Look for recently modified files, unfamiliar files/folders, or suspicious code injected into your existing files (especially `index.php`, `.htaccess`, theme files, plugin files). Common injections include obfuscated PHP, JavaScript, or iframes.
    *   **Database Check:** For spam injections, check your database tables for unexpected content.

3.  **Clean Your Site:**
    *   **Remove Malicious Code/Files:** Delete any files or code identified as malicious. If core CMS files are compromised, replace them with fresh copies from the official source (e.g., WordPress.org). *Do not upload your own potentially compromised backups unless you are certain they are clean from before the hack.*
    *   **Remove Hacked Content:** Delete spammy pages or content injected into your database.
    *   **Restore from a Clean Backup (If Available):** If you have a known clean backup from *before* the hack occurred, restoring it can be the quickest way to clean the site. However, you *must* then immediately patch the vulnerability that allowed the hack.

4.  **Fix the Vulnerability:**
    *   This is crucial. If you only clean the site but don't fix how the attackers got in, you'll likely be hacked again.
    *   **Common Vulnerabilities:**
        *   Outdated software (CMS like WordPress, Joomla, Drupal; plugins; themes). **Update everything to the latest versions.**
        *   Weak passwords (admin, FTP, database, hosting cPanel). **Change all passwords to strong, unique ones.**
        *   Insecure themes or plugins (especially nulled/pirated ones). Remove them.
        *   Server misconfigurations.
        *   Compromised user accounts.

5.  **Secure Your Site (Proactive Measures):**
    *   **Implement a Web Application Firewall (WAF):** Services like Cloudflare or Sucuri WAF can help block malicious traffic.
    *   **Regularly Update Everything:** CMS, plugins, themes.
    *   **Use Strong Passwords & 2FA:** For all accounts.
    *   **Limit User Permissions:** Only give users the access they need.
    *   **Regular Backups:** Store them off-site.
    *   **Install Security Plugins (for CMS like WordPress):** E.g., Wordfence, Sucuri Security.
    *   **File Integrity Monitoring:** To detect unauthorized changes.

6.  **Request a Review in Google Search Console:**
    *   Once you are *certain* your site is clean and the vulnerability is fixed:
        *   Go to GSC > Security & Manual Actions > Security Issues.
        *   Click the button to "Request Review."
        *   Briefly explain the steps you took to clean the site and fix the vulnerability.
    *   Google will re-scan your site. If it's clean, they will remove the security warnings (this can take a day or two).

**Example Scenario: Hacked WordPress site injecting spammy links.**
1.  **Notification:** GSC shows "Hacked: Spam" with example URLs like `www.yoursite.com/buy-cheap-viagra-online`.
2.  **Isolate:** Put site in maintenance mode.
3.  **Identify:** Install Wordfence, run a scan. It identifies a recently uploaded malicious PHP file in `wp-content/uploads/` and modified `wp-config.php`.
4.  **Clean:** Delete the malicious file. Restore `wp-config.php` from a clean backup or manually remove injected code. Scan database for spammy posts/pages and remove them.
5.  **Fix Vulnerability:** Discover an outdated plugin was the entry point. Update all plugins, themes, and WordPress core. Change all admin, FTP, and database passwords.
6.  **Secure:** Install and configure Wordfence for ongoing protection.
7.  **Request Review:** Submit review request in GSC explaining actions taken.
